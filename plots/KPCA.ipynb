```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate Gaussian surface
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2)/2)

# Generate random points on the Gaussian surface
np.random.seed(42)
num_points = 300
x_pts = np.random.normal(0, 1.2, num_points)
y_pts = np.random.normal(0, 1.2, num_points)
z_pts = np.exp(-(x_pts**2 + y_pts**2)/2)

# Add small vertical noise for visibility
z_pts += np.random.normal(0, 0.03, num_points)

# Filter points based on vertical position
mask = (z_pts > 0.55) | (z_pts < 0.4)
x_filtered = x_pts[mask]
y_filtered = y_pts[mask]
z_filtered = z_pts[mask]

# Assign colors to filtered points
colors = np.where(z_filtered > 0.5, 'red', 'green')

# Create horizontal separating plane at z=0.5
plane_x = np.linspace(-3, 3, 20)
plane_y = np.linspace(-3, 3, 20)
Plane_X, Plane_Y = np.meshgrid(plane_x, plane_y)
Plane_Z = np.full_like(Plane_X, 0.5)  # Constant height plane

# Create plot
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.grid(False)

# Plot colored points
ax.scatter(x_filtered, y_filtered, z_filtered, c=colors, s=30, depthshade=True)

# Plot horizontal decision surface
ax.plot_surface(Plane_X, Plane_Y, Plane_Z, alpha=0.5, color='grey')

ax.view_init(elev=8, azim=-45)  # Adjust viewing angle
plt.savefig('3d_KPCA.png',format="png",bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set number of points and spread
n_red = 30      # Number of red points
n_green = 20    # Number of green points
spread = 0.1    # Controls clustering tightness (smaller = tighter cluster)

# Generate clustered coordinates using normal distribution
red_x = np.random.normal(0.5, spread, n_red)
red_y = np.random.normal(0.5, spread, n_red)
green_x = np.random.normal(0.5, spread, n_green)
green_y = np.random.normal(0.5, spread, n_green)

# Create plot
plt.figure(figsize=(8, 6))
plt.scatter(red_x, red_y, color='red', label='Red Points', alpha=0.7)
plt.scatter(green_x, green_y, color='green', label='Green Points', alpha=0.7)

# Customize axes and borders
ax = plt.gca()
ax.spines[['right', 'top']].set_visible(False)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(f'Tighter Clusters (Spread: {spread})')

# Set plot limits and grid
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True, alpha=0.2)
plt.legend()

plt.show()
```

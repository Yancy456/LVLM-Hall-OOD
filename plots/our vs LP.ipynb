```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['VQA v2', 'POPE', 'VizWiz', 'VQA2v-IDK']
AUROC_LP = [82.00, 91.75, 88.00, 84.31]
AUROC_ours = [79.51, 85.35, 72.00, 72.55]

# Bar width
bar_width = 0.3

# Positions of the bars on the x-axis
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(7.5, 5))
# Creating the bar chart
plt.bar(r1, AUROC_LP, width=bar_width, label='AUROC of LP',color='tab:blue')
plt.bar(r2, AUROC_ours, width=bar_width, label='AUROC of Ours',color='tab:orange')

# Adding labels and title
plt.ylabel('Values')
plt.xticks([r + bar_width/2 for r in range(len(categories))], categories)

# Adding legend
plt.legend(ncol=2,loc='upper right')

plt.savefig('LP_vs_ours.pdf',format="pdf",bbox_inches='tight')
# Show the plot
plt.show()
```

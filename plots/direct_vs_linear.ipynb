```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['VQA v2', 'POPE', 'VizWiz', 'VQA2v-IDK']
acc_DP = [66.48, 40.75, 65.30, 68.25]
acc_ours = [81.40, 83.35, 68.00, 72.55]
AUROC_DP = [80.40, 90.03, 72.38, 71.73]
AUROC_ours = [79.38, 89.86, 70.11, 78.31]

# Bar width
bar_width = 0.15

# Positions of the bars on the x-axis
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
plt.figure(figsize=(7.5, 5))
# Creating the bar chart
plt.bar(r1, acc_DP, width=bar_width, label='Acc of DP',color='tab:blue',hatch='//')
plt.bar(r2, acc_ours, width=bar_width, label='Acc of Ours',color='tab:blue')
plt.bar(r3, AUROC_DP, width=bar_width, label='AUROC of DP',color='tab:orange',hatch='//')
plt.bar(r4, AUROC_ours, width=bar_width, label='AUROC of Ours',color='tab:orange')

# Adding labels and title
plt.ylabel('AUROC')
plt.xticks([r + 1.5 * bar_width for r in range(len(categories))], categories)

# Adding legend
plt.legend(ncol=2,loc='upper center', bbox_to_anchor=(0.70, 1))

plt.savefig('direct_vs_linear.pdf',format="pdf",bbox_inches='tight')
# Show the plot
plt.show()
```
